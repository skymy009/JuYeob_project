```python
# 학원 데이터 가져오기
data_academy = data[(data['시도명'] == '부산광역시') & (data['상권업종대분류명'] == '학문/교육')].copy()
data_academy.head()
```

```python
academy_count = data_academy['상권업종소분류명'].value_counts().head(30)
academy_count
```

```python
# 빈도수가 1000개 이상인 데이터만 따로 본다.
academy_count_1000 = academy_count[academy_count > 1000]
academy_count_1000
```

## Folium 으로 지도 활용하기

```python
import folium 
```

```python
long = data_academy_selected['경도'].mean()
lat = data_academy_selected['위도'].mean()
```

```python
data_m = data_academy_selected[data_academy_selected['상권업종소분류명'].isin(['어린이집' , '학원-입시'])]
data_m = data_m.sample(1000)
data_m.shape
```

```python
data_m.iloc[0]
```

```python
m = folium.Map(location=[lat, long], zoom_start=12)
folium.Marker([37.5545, 126.867], tooltip='해법수학').add_to(m)
m
```

```python
m.save('index.html')
```

```python
data_m.index
```

```python
for i in data_m.index[:10]:
    tooltip = data_m.loc[i, "상호명"] +"-"+ data_m.loc[i, "도로명주소"]
    lat = data_m.loc[i, "위도"]
    long = data_m.loc[i, "경도"]
    
    print(tooltip, lat, long)
```

```python
m = folium.Map(location=[lat, long], zoom_start=12, 
    tiles='Stamen Toner')

for i in data_m.index[:10]:
    tooltip = data_m.loc[i, "상호명"] +"-"+ data_m.loc[i, "도로명주소"]
    lat = data_m.loc[i, "위도"]
    long = data_m.loc[i, "경도"]
    
    folium.CircleMarker([lat, long], tooltip=tooltip, radius=3).add_to(m)
m
```
