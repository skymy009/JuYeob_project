```python
data20 = pd.concat([data2001, data2002, data2003, data2004, data2005, data2006, data2007, data2008,
                  data2009, data2010, data2011, data2012], ignore_index = True)
print(data20.head())
```

```python
data20 = data20[data20['측정소명'].str.contains('학장동')]
data20.reset_index(drop = True, inplace=True)
```

```python
data20.head()
```

```python
data20 = data20[["측정일시", "PM10", "PM25", "SO2", "CO", "NO2", "O3"]]
```

```python
data20.head()
```

```python
data20['Datetime'] = data20["측정일시"].astype(str)
data20['Datetime'] = data20["Datetime"].str[8:]
data20["Date"] = data20["측정일시"].astype(str)
data20["Date"] = data20["Date"].str[:8]
```

```python
data20.head()
```

```python
data20['Date'] = pd.to_datetime(data20['Date'])
data20['Datetime'] = data20['Datetime'].astype(float)
```

```python
data20.head()
```

```python
for i in range(len(data20)):
    data20['Date'].iloc[i] = data20['Date'].iloc[i] + timedelta(hours = data20['Datetime'].iloc[i]) 
data20.set_index('Date', inplace = True)
```

```python
data20.head()
```

```python
data20 = data20.asfreq('H')
```

```python
data20.index
```

```python
data20.isnull().sum()
```

```python
data20.asfreq('H', method = 'ffill')
data20.fillna(method='ffill', inplace = True)
data20.fillna(method='bfill', inplace = True)
```

```python
data20.isnull().sum()
```

```python
data20.head()
```

```python
data20 = data20[["PM10","PM25","SO2","CO","NO2","O3"]]
```

```python
data20[['PM10']].plot(kind = 'line', figsize = (15,6), linewidth = 3, fontsize = 20)
data20[['PM25']].plot(kind = 'line', figsize = (15,6), linewidth = 3, fontsize = 20)
data20[['SO2']].plot(kind = 'line', figsize = (15,6), linewidth = 3, fontsize = 20)
data20[['NO2']].plot(kind = 'line', figsize = (15,6), linewidth = 3, fontsize = 20)
data20[['CO']].plot(kind = 'line', figsize = (15,6), linewidth = 3, fontsize = 20)
data20[['O3']].plot(kind = 'line', figsize = (15,6), linewidth = 3, fontsize = 20)
plt.xlabel('Date', fontsize=13)
plt.ylabel('thing', fontsize=13)
plt.show()
```
