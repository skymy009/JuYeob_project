```python
data = pd.concat([data16,data17,data18,data19,data20], ignore_index = True)
data.head()
```

```python
data = data[["측정일시", "PM10", "PM25", "SO2", "CO", "NO2", "O3"]]
data.head()
```

```python
#결측치 확인
data.isnull().sum()
```

```python
data.describe()
```

```python
#결측치를 각 수치들의 평균으로 대체함.
data["PM10"].fillna(data["PM10"].mean(), inplace = True)
data["PM25"].fillna(data["PM25"].mean(), inplace = True)
data["SO2"].fillna(data["SO2"].mean(), inplace = True)
data["CO"].fillna(data["CO"].mean(), inplace = True)
data["NO2"].fillna(data["NO2"].mean(), inplace = True)
data["O3"].fillna(data["O3"].mean(), inplace = True)
```

```python
#결측치가 없음을 확인
data.isnull().sum()
```

```python
#측정일시를 문자열로 변환하고 1~8번째 문자로 대체한다.
data['측정일시'] = data['측정일시'].astype(str)
data['측정일시'] = data['측정일시'].str[:8]
data.head()
```

```python
#datatime을 사용하여 날짜 형식으로 변환한다.
data["측정일시"] = pd.to_datetime(data["측정일시"])
data.head()
```

```python
data.info()
```

```python
#피어슨 상관계수 분석
#각 변수 간의 상관성을 확인한다.
corr = data.corr(method = 'pearson')
```

```python
corr
```

```python
#corr결과 : ±0.3 이상일 경우 약한 상관성이 있다.
#           ±0.7 이상일 경우 강한 상관성이 있다.
#PM10과 PM25는 강한 상관관계임을 알 수 있다.
```

```python
#선형 회귀분석
#종속변수 : PM25(미세먼지)
#독립변수 : PM10(초미세먼지), SO2(황산화물), C0(일산화탄소), NO2(질소산화물), O3(오존)
```

```python
from sklearn.model_selection import train_test_split
x_data = data[['PM10','SO2','CO', 'NO2', 'O3']]
target = data[['PM25']]
x = x_data
y = target
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)
```

```python
print(len(x_train), len(x_test), len(y_train), len(y_test))
```

```python
# 선형 회귀 분석 모형 객체 생성
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# train data를 가지고 모형 학습
lr.fit(x_train, y_train)
```

```python
# 학습을 마친 모형에 test data를 적용하여 결정계수(R제곱) 계산
r_square = lr.score(x_test, y_test)
r_square
```

```python
y_predict = lr.predict(x_test)
```

```python
plt.scatter(y_test, y_predict, alpha=0.4)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("MULTIPLE LINEAR REGRESSION")
plt.show()
```
